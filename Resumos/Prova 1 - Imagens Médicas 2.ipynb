```python
def histogram(img):
    N, M = img.shape
    n = N*M
    bins, hist = np.unique(img.flatten(), return_counts=True)
    
    fig, ax = plt.subplots(1,2)
    
    ax[0].stem(bins, hist, linefmt='-k', markerfmt=' ', basefmt=' ')
    ax[1].imshow(img, cmap='gray')
    #ax[1].stem(bins/n, hist/n, linefmt='-k', markerfmt='k', basefmt='')

    ax[0].set_title('Histogram')
    ax[1].set_title('Image')

    ax[0].set_xlabel('Levels')
    ax[0].set_ylabel('Counting of levels')
    
    ax[1].grid(False)
    ax[1].set_xticks([])
    ax[1].set_yticks([])
    
    fig.set_size_inches(10,4)

    print('Média = ' + str(img.mean()))
    print('Desvio padrão = ' + str(img.std()))
    
    return bins, hist
```

## Questão 3

Ao aplicar as diversas transformações de intensidade, os efeitos para cada uma das transformações na imagem seria:

- Função logarítmica:

O uso de funções logarítmicas permite mapear uma faixa estreita de baixos níveis de entrada a uma ampla faixa de níveis de saída, permitindo que detalhes antes escuros tornem-se mais claros e que pixels de intensidade já elevada não sofram muito efeito. Nesse caso detalhes dos tecidos dos órgãos ou vasos sanguíneos possuirão mais brilho.

- Função exponencial:

Funções exponenciais se comportam de acordo com o fator gamma ajustado, de forma que um baixo fator gamma permitirá um comportamento semelhante à função logarítmica, em que uma faixa estreita de baixos níveis de entrada sejam mapeados em uma faixa mais larga de níveis de saída. Mas de acordo com que o fator gamma é aumentado e quando supera o valor unitário esse comportamento se inverte, permitindo reduzir esse efeito, até que para um gamma alto a função mapeará uma larga faixa de níveis de entrada a uma estreita e de baixos níveis de valores de saída. Para casos que o gamma seja maior que a unidade, o efeito na imagem será o escurecimento generalizado dos pixels, com apenas alguns níveis mais altos e que superam o mapeamento da faixa inicial de entrada ainda se destacando com leves tons de cinza. 

- Negativo:

O efeito do negativo é muito simples, uma vez que consiste no complemento dos valores dos pixels, resultando na inversão das cores mais claras pelas cores escuras de acordo com a intensidade de cada pixel original. Os ossos que aparecem em branco ficariam escuros e o fundo escuro tornaria mais claro.
